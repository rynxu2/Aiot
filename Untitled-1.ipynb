```python
import numpy as np
import matplotlib.pyplot as plt

def plot_activity(activity, data):
    fig, ax = plt.subplots(figsize=(18, 4))

    # Vẽ dữ liệu gia tốc trên ba trục
    plt.plot(data['Time'], data['AccelX'], color='orchid', linewidth=1.5, label="AccelX")
    plt.plot(data['Time'], data['AccelY'], color='dodgerblue', linewidth=1.5, label="AccelY")
    plt.plot(data['Time'], data['AccelZ'], color='dimgrey', linewidth=1.5, label="AccelZ")

    # Điều chỉnh biểu đồ
    plt.subplots_adjust(hspace=0.2, top=0.90)
    plt.suptitle(f"Activity: {activity}", fontsize=14)
    plt.legend()
    plt.ylim(-25, 25)
    plt.grid(True)
    plt.xticks([])
    plt.show()

# Duyệt qua từng hoạt động và vẽ biểu đồ
for activity in np.unique(df['ActivityLabel']):
    subset = df[df['ActivityLabel'] == activity][:200]
    plot_activity(activity, subset)
```

```python
for activity in np.unique(df['ActivityLabel']):
    subset = df[df['ActivityLabel'] == activity][200:400]
    plot_activity(activity, subset)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_activity_gyro(activity, data):
    fig, ax = plt.subplots(figsize=(18, 4))

    # Vẽ dữ liệu gia tốc trên ba trục
    plt.plot(data['Time'], data['GyroX'], color='orchid', linewidth=1.5, label="GyroX")
    plt.plot(data['Time'], data['GyroY'], color='dodgerblue', linewidth=1.5, label="GyroY")
    plt.plot(data['Time'], data['GyroZ'], color='dimgrey', linewidth=1.5, label="GyroZ")

    # Điều chỉnh biểu đồ
    plt.subplots_adjust(hspace=0.2, top=0.90)
    plt.suptitle(f"Activity: {activity}", fontsize=14)
    plt.legend()
    plt.ylim(-25, 25)
    plt.grid(True)
    plt.xticks([])
    plt.show()

# Duyệt qua từng hoạt động và vẽ biểu đồ
for activity in np.unique(df['ActivityLabel']):
    subset = df[df['ActivityLabel'] == activity][:200]
    plot_activity_gyro(activity, subset)
```
